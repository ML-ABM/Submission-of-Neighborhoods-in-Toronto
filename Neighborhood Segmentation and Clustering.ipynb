```python
import pandas as pd
import wikipedia as wp
from geopy.geocoders import Nominatim 
```

## Loading of the table from Wikipedia 

```python
html = wp.page("List_of_postal_codes_of_Canada:_M").html().encode("UTF-8")
df1 = pd.read_html(html)[0]
```

```python
df1.head()
```

```python
df1.shape
```

##  Data Wrangling

```python
df=df1.drop(0)
df=df[df.Borough !='Not assigned']
```

```python
df.head()
```

```python
df[df.Borough =='Not assigned'].count()
```

```python
df.shape
```

## Loading the geographical coordinates

```python
postal_data =pd.read_csv('http://cocl.us/Geospatial_data')
```

```python
postal_data.head()
```

##  Merging of the two tables

```python
df_all=pd.merge(df, postal_data, on='Postal Code')
```
